# FAST EDA (Análisis Exploratorio Rápido)

```python
# Instalar librerías necesarias
!pip install ydata-profiling --quiet

import pandas as pd
import os
from ydata_profiling import ProfileReport
from google.colab import files
from IPython.display import display
import io

# Subir archivo
uploaded = files.upload()

# Detectar el archivo subido
filename = next(iter(uploaded))

# Detectar tipo de archivo y leer
if filename.endswith('.csv'):
    df = pd.read_csv(io.BytesIO(uploaded[filename]))
elif filename.endswith(('.xls', '.xlsx')):
    df = pd.read_excel(io.BytesIO(uploaded[filename]))
else:
    raise ValueError("Formato no soportado. Por favor sube un archivo CSV o Excel.")

# Mostrar primeras filas (opcional)
display(df.head())

# Generar perfil
profile = ProfileReport(df, title="Reporte Exploratorio", explorative=True)

# Guardar como HTML
output_filename = "reporte_exploratorio.html"
profile.to_file(output_file=output_filename)

# Descargar automáticamente
files.download(output_filename)
```
